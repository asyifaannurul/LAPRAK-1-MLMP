##### *Features* *Engineering* adalah langkah-langkah untuk mengubah data mentah menjadi fitur-fitur yang dapat digunakan dalam algoritma *machine* *learning*. Selama proses *feature* *engineering*, variabel atau fitur yang paling berpengaruh dipilih dan direkayasa untuk pemodelan. *Feature* engineering memerlukan pemahaman yang baik karena melibatkan kombinasi analisis data, pengetahuan tentang data, dan sedikit intuisi. Tujuan dari *feature* *engineering* adalah menyederhanakan dan mempercepat transformasi data sekaligus meningkatkan akurasi model. kualitas fitur jauh lebih penting daripada seberapa canggih algoritma  yang dipilih. ### A. *Handling Missing Value*

>*Missing values* atau *missing data*, terjadi ketika tidak ada data /
tidak ada nilai
yang tersimpan untuk observasi tertentu dalam suatu variabel. *Missing* data adalah
kejadian yang umum dan dapat berdampak signifikan pada pemodelan *machine* *learning*. Data yang tidak lengkap adalah masalah yang tidak terhindarkan dalam
menangani sebagian besar sumber data.
### B. Kardinalitas
>Kardinalitas merujuk pada jumlah nilai unik, atau label, yang terdapat dalam
suatu variabel kategoris. Konsep ini digunakan untuk mengukur granularitas atau
keunikan data dalam sebuah fitur. Kardinalitas dapat diklasifikasikan dalam spektrum, namun umumnya dibedakan menjadi kardinalitas rendah (*low cardinality*) dan kardinalitas tinggi (*high cardinality*). Pengaruh kardinalitas, sangat krusial terhadap performa model dan efisiensi komputasi. Pertama, variabel high cardinality dapat menyebabkan curse of dimensionality  jika ditransformasi
menggunakan metode encoding standar seperti One-Hot Encoding. Metode ini akan menciptakan banyak fitur biner baru (sesuai jumlah label), membuat data menjadi sangat jarang dan secara drastis meningkatkan kompleksitas komputasi. Kedua, kardinalitas secara signifikan meningkatkan risiko overfitting. Hal ini terjadi karena
banyak label mungkin hanya muncul beberapa kali dalam set data latih, sehingga model cenderung menghafal pola spesifik yang terkait dengan label langka tersebut, alih-alih mempelajari pola umum yang dapat digeneralisasi ke data yang belum pernah terlihat.
### C. *Splitting* Data
>Splitting data merupakan strategi fundamental dalam machine learning
untuk membagi himpunan data menjadi bagian terpisah: data training dan data
testing. Pembagian ini krusial untuk membangun model yang robust dan dapat
digeneralisasi dengan baik pada data baru (Muraina, 2022). Proporsi pembagian
data bersifat subjektif dan fleksibel. Rasio yang umum digunakan adalah 80% untuk
training dan 20% untuk testing (Woschnagg dan Cipan, 2004). Namun, seperti yang
ditekankan oleh Géron (2019), rasio ini dapat disesuaikan tergantung pada ukuran
dataset.
### D. Handling Outlier
>Outlier, atau dikenal sebagai pencilan, merupakan observasi atau titik data
yang menunjukkan deviasi ekstrem dan berbeda secara signifikan dari sebagian
besar data lain dalam suatu set data. Keberadaan outlier dapat diatribusikan ke
berbagai sumber, mulai dari kesalahan pengukuran, kesalahan entri data (misinput),
kontaminasi data, hingga representasi sah dari kejadian langka atau variabilitas
inheren dalam populasi. Dalam analisis statistik dan pemodelan machine learning,
outlier menuntut perhatian khusus karena memiliki potensi untuk memberikan
pengaruh yang tidak proporsional (disproportionate influence) terhadap hasil.
### E. Scaling data
>Scaling data adalah salah satu langkah pra-pemrosesan dalam analisis data
untuk mentransformasi variabel data ke dalam rentang skala yang sama. Tujuannya
adalah untuk memastikan bahwa tidak ada satu variabel independen pun yang
mendominasi proses pembelajaran hanya karena memiliki rentang nilai yang lebih
besar dibandingkan variabel independen lainnya. Sebagaimana dinyatakan oleh
Singh dan Singh (2020), normalisasi data sering kali dapat meningkatkan performa
model prediktif secara signifikan, terutama untuk algoritma yang sensitif terhadap
skala input.

### F. Encoding
>Feature encoding adalah proses mengubah fitur kategoris atau non-numerik
menjadi format numerik yang dapat digunakan sebagai input untuk algoritma
machine learning. Banyak algoritma machine learning membutuhkan input
numerik. Ada beberapa teknik umum untuk feature encoding, seperti One-Hot
Encoding dan Ordinal Encoding.

## G. Imbalanced Dataset

kategorinya tidak setara. Kelas-kelas yang mencakup proporsi besar dari set data

### D. *Handling Outlier*
>*Handling outlier* berguna untuk meningkatkan keandalan hasil analisis dan model, memastikan bahwa data yang digunakan mewakili kondisi sebenarnya tanpa distorsi dari nilai-nilai ekstrem yang tidak wajar.

```python
#buat list fitur numerik
list_num = ['Suhu Udara', 'Kelembapan', 'Kecepatan Angin']
# Hitung IQR
list_outlier=[]
list_lower_bound =[]
list_upper_bound =[]
for i in X_train[list_num]:
    Q1 = np.percentile(X_train[i], 25)
    Q3 = np.percentile(X_train[i], 75)
    IQR = Q3 - Q1

    # Tentukan batas bawah dan batas atas untuk outlier
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    list_lower_bound.append(lower_bound)
    list_upper_bound.append(upper_bound)

    # Hitung jumlah outlier di bawah batas bawah dan di atas batas atas
    num_outliers_lower = np.sum(X_train[i] < lower_bound)
    num_outliers_upper = np.sum(X_train[i] > upper_bound)

    # Jumlah total outlier
    total_outliers = num_outliers_lower + num_outliers_upper
    list_outlier.append(total_outliers)

#Mendefinikan dataframe baru mengenai outliers
outliers = pd.DataFrame()
outliers['Kolom'] = list_num
outliers['Jumlah Outlier'] = list_outlier
outliers['Lower Bound'] = list_lower_bound
outliers['Upper Bound'] = list_upper_bound
outliers
```

>Pada tahap ini dilakukan identifikasi terhadap adanya *outlier* pada variabel numerik. Variabel yang digunakan adalah Suhu Udara, Kelembapan, dan Kecepatan Angin.Berdasarkan hasil deteksi dengan metode IQR, hanya variabel Kecepatan Angin yang mengandung *outlier*.

```python
#mendefinisikan kolom yang digunakan dan list tempat value skewness
df_num = X_train[list_num]
nilai_skew = []
nilai_skew_normal = []

#membuat fungsi untuk memetakan distribusi setiap kolom
for i in df_num.columns:
    skewness = X_train[i].skew()
    if -0.5 <= skewness <= 0.5:
        nilai_skew_normal.append(i)
    else:
        nilai_skew.append(i)

# Menampilkan list
print('kolom yang mempunyai nilai skewness sedang:', nilai_skew)
print('kolom yang mempunyai nilai skewness normal:', nilai_skew_normal)
```

>Analisis skewness dilakukan untuk mengetahui bentuk distribusi data pada setiap variabel numerik, apakah simetris (mendekati distribusi normal) atau condong ke kiri/kanan.
Nilai skewness diinterpretasikan sebagai berikut:
>1.   -0,5 ≤ Skewness ≤ 0,5 → Distribusi normal (simetris)
2.   Skewness < -0,5 atau Skewness > 0,5 → Distribusi miring (skewed)

>Berdasarkan analisis skewness yang telah dilakukan diketahui bahwa variabel Suhu Udara memiliki nilai skewness dalam rentang normal, yang berarti distribusinya relatif simetris sedangkan variabel Kelembapan dan Kecepatan Angin memiliki nilai skewness di luar rentang normal, sehingga datanya miring (tidak simetris). Hal ini menunjukkan adanya konsentrasi data di satu sisi distribusi.

####      1. Capping
>Capping adalah metode penanganan *outlier* dengan cara membatasi nilai ekstrem agar tidak melewati batas tertentu, tanpa menghapus data.

```python
# Capping SEMUA kolom numerik
winsoriser_all = Winsorizer(
    capping_method='iqr',
    tail='both',
    fold=1.5,
    variables=['Suhu Udara', 'Kelembapan', 'Kecepatan Angin'],  # ← SEMUA KOLOM
    missing_values='ignore'
)

X_train_capped = winsoriser_all.fit_transform(X_train)
X_test_capped = winsoriser_all.transform(X_test)
```

```python
# Plotting histogram and boxplot before and after capping
for col in list_num:
  print(f'{col} - Before Capping')
  diagnostic_plots(X_train, col)

  print(f'\n{col} - After Capping')
  diagnostic_plots(X_train_capped, col)
```

Prosesnya dilakukan dengan fit_transform pada data train dan transform pada data test, lalu hasilnya disimpan sebagai X_train_scale dan X_test_scale untuk tahap modeling selanjutnya.

### F. Encoding
>Encoding berguna untuk mengonversi data kategorik menjadi numerik, agar model dapat membaca dan mempelajari pola dari data tersebut secara matematis

####     One Hot Encoder
>Mengubah setiap kategori menjadi kolom baru bernilai 0 atau 1.

```python
list_cat = ['Keadaan_Cuaca_reduced']

encoder = OneHotEncoder()

#melakukan encoding
X_train_encoded = encoder.fit_transform(X_train_scale[list_cat])
X_test_encoded = encoder.transform(X_test_scale[list_cat])
```

>Karena kolom Keadaan_Cuaca_reduced merupakan variabel kategorikal, proses One-Hot Encoding dilakukan menggunakan OneHotEncoder untuk mengubah kategori menjadi bentuk numerik biner. Encoder di-fit dan di-transform pada data train agar model mempelajari struktur kategorinya, kemudian hanya di-transform pada data test untuk menjaga konsistensi. Hasil encoding disimpan sebagai X_train_encoded dan X_test_encoded dan akan digabungkan dengan fitur numerik hasil scaling sebelum masuk ke tahap modeling.

### G. Balancing Data
>Balancing data berguna untuk mengurangi ketimpangan jumlah antar kelas, sehingga model klasifikasi menjadi lebih adil, akurat, dan tidak bias terhadap kelas dominan.

```python
y_train.value_counts()
```

```python
# Oversampling with SMOTE

smote = SMOTE(sampling_strategy='minority', k_neighbors=5, random_state=42)     # Minority Class will have same the same number as the majority class
# smote = SMOTE(sampling_strategy={1:1000}, k_neighbors=5, random_state=42)     # Minority Class will have 1000 data
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_encoded, y_train)
y_train_balanced.value_counts()
```

>SMOTE digunakan untuk menangani ketidakseimbangan kelas dengan cara membuat sampel sintetis pada kelas minoritas. Dengan sampling_strategy='minority', jumlah data minoritas dibuat sama dengan mayoritas. Proses fit_resample menghasilkan X_train_balanced dan y_train_balanced yang kini memiliki distribusi kelas seimbang sehingga model tidak bias.

## **Training Model**

>Training model adalah proses melatih algoritma machine learning agar dapat mengenali pola dari data, sehingga nantinya bisa memprediksi atau mengklasifikasikan data baru dengan akurat. Pada tahap pemodelan, algoritma yang digunakan adalah Logistic Regression, yaitu metode klasifikasi linier yang umum digunakan untuk memprediksi variabel target kategorikal. Model ini dipilih karena sederhana, cepat dilatih, serta mampu memberikan interpretasi yang jelas melalui analisis koefisien fitur. Proses pelatihan dilakukan menggunakan data train yang telah dipersiapkan pada setiap skenario—baik tanpa handling, dengan handling (encoded & scaled), maupun dengan data yang sudah di-handling dan diseimbangkan menggunakan SMOTE—agar dapat dibandingkan pengaruh setiap tahapan preprocessing terhadap performa model.

```python
from sklearn.linear_model import LogisticRegression
```

```python
#tanpa feature engineering
lr_1 = LogisticRegression()
lr_1.fit(X_train, y_train)
```

```python
#dengan feature engineering
lr_2 = LogisticRegression()
lr_2.fit(X_train_encoded, y_train)
```

```python
#dengan feature engineering dan balancing data
lr_3 = LogisticRegression()
lr_3.fit(X_train_balanced, y_train_balanced)
```

## **Model Performance**


> Sebagai bagian dari proses penilaian performa model, evaluasi dilakukan dalam tiga tahap berbeda. Pertama, model diuji menggunakan data asli tanpa proses handling, sehingga terlihat bagaimana performanya sebelum dilakukan preprocessing apa pun. Kedua, model dievaluasi menggunakan data yang sudah melalui proses handling, termasuk encoding dan scaling, untuk melihat peningkatan akurasi setelah transformasi fitur. Ketiga, evaluasi dilakukan menggunakan data yang sudah di-encode dan di-handling sekaligus diseimbangkan (balanced) menggunakan SMOTE, guna menilai sejauh mana penanganan ketidakseimbangan kelas dapat memperbaiki kemampuan model dalam mengklasifikasi kedua kelas secara lebih adil.



```python
from sklearn.metrics import classification_report
```

```python
# Check Results - without Handling

print(classification_report(y_train, lr_1.predict(X_train)))
print(classification_report(y_test, lr_1.predict(X_test)))
```

>Hasil evaluasi tanpa penanganan imbalance menunjukkan bahwa model sangat bias ke kelas mayoritas (kelas 2). Hal ini terlihat dari recall kelas 1 yang sangat rendah (11% pada train dan 32% pada test), menandakan model gagal mengenali sebagian besar data kelas minoritas.

```python
# Check Results - with Handling (Encoded)

print(classification_report(y_train, lr_2.predict(X_train_encoded)))
print(classification_report(y_test, lr_2.predict(X_test_encoded)))
```

>Hasil evaluasi setelah dilakukan handling meliputi encoding fitur kategorikal dan scaling pada fitur numerik menunjukkan adanya peningkatan performa dibandingkan kondisi tanpa handling, terutama pada kemampuan model mengenali kelas minoritas. Hal ini terlihat dari naiknya recall kelas 1 baik pada data train (dari 11% menjadi 31%) maupun pada data test (dari 32% menjadi 42%).

```python
# Check Results - with Handling (Balanced)
print(classification_report(y_train_balanced, lr_3.predict(X_train_balanced)))

# Check Results - with Handling (Encoded)
print(classification_report(y_test, lr_3.predict(X_test_encoded)))
```

>Evaluasi model Logistic Regression dilakukan dalam tiga tahap: tanpa handling, dengan handling (encoded & scaled), dan dengan handling + balancing (SMOTE). Pada tahap tanpa handling, model sangat bias ke kelas mayoritas dengan akurasi 0.77 (train) dan 0.81 (test), tetapi recall kelas minoritas sangat rendah. Setelah handling, performa kelas minoritas membaik—recall naik dan macro average meningkat—dengan akurasi tetap stabil di sekitar 0.79–0.81. Setelah dilakukan balancing menggunakan SMOTE, model menjadi lebih seimbang dalam mengenali kedua kelas. Akurasi train menjadi 0.64 (wajar karena kelas kini seimbang), dan akurasi test tetap 0.81, namun recall kelas minoritas meningkat menjadi 55%. Secara keseluruhan, preprocessing dan balancing membuat model lebih adil pada kedua kelas tanpa mengorbankan performa pada data uji.

## **Kesimpulan**


> Feature engineering adalah proses mengubah, memperbaiki, atau menambah fitur agar model machine learning dapat memahami pola data dengan lebih baik. Konsep ini mencakup beberapa jenis, seperti data cleaning (mengatasi missing value, outlier), feature transformation (scaling, encoding, normalisasi, log transform), feature creation (membuat variabel baru dari kombinasi fitur lama), dimensionality reduction (PCA, SVD), hingga feature selection untuk memilih fitur paling informatif. Tujuannya adalah meningkatkan kualitas representasi data sehingga model dapat bekerja lebih optimal dan memiliki performa prediksi yang lebih baik.

>Di Python, feature engineering umumnya dilakukan menggunakan library seperti pandas, NumPy, dan pustaka preprocessing dari scikit-learn. Tekniknya meliputi pembersihan data (drop atau imputasi nilai hilang), transformasi nilai numerik seperti scaling dengan StandardScaler, MinMaxScaler, atau RobustScaler, encoding variabel kategorikal menggunakan OneHotEncoder, LabelEncoder, atau OrdinalEncoder, serta balancing kelas menggunakan SMOTE dari imblearn. Semua teknik ini dapat digabung dalam pipeline preprocessing sehingga proses pengolahan data menjadi lebih terstruktur dan konsisten sebelum model dilatih.

>Dari hasil feature engineering dan evaluasi model Logistic Regression dengan tiga skenario—tanpa handling, handling (encoding & scaling), dan handling + balancing (SMOTE)—diperoleh kesimpulan bahwa preprocessing memberikan dampak besar pada performa model. Tanpa handling, model sangat bias terhadap kelas mayoritas meskipun akurasinya tampak tinggi. Setelah dilakukan handling, performa kelas minoritas mulai membaik. Tahap terbaik diperoleh setelah penerapan SMOTE, di mana model menjadi lebih seimbang dalam mengenali kedua kelas: akurasi test tetap stabil di 0.81, tetapi recall kelas minoritas meningkat menjadi 55% sehingga klasifikasi kedua kelas lebih adil dan representatif. Ini menunjukkan bahwa kombinasi capping, RobustScaler, OneHotEncoder, dan SMOTE menghasilkan model yang jauh lebih optimal dan tidak bias.
